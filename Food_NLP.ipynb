```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize SentenceTransformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Function to encode sentences
def encode_sentences(sentences):
    return model.encode(sentences)

# Function to process large CSV in chunks
def process_large_csv(file_path, chunk_size=10000):
    reader = pd.read_csv(file_path, chunksize=chunk_size)
    sentences = []
    for chunk in reader:
        sentences.extend(chunk['name'].tolist())

    # Encode all sentences from the CSV
    encoded_sentences = encode_sentences(sentences)
    print(encoded_sentences)

    # Prompt for input sentence
    input_sentence = input("Enter a sentence to find similar sentences: ")

    # Encode the input sentence
    encoded_input = encode_sentences([input_sentence])[0]
    print(encoded_input)

    # Compute pairwise cosine similarities between input and CSV sentences
    similarities = cosine_similarity([encoded_input], encoded_sentences)
    print(type(similarities))
    print(similarities[0])

    # Example: Find top similar sentences
    k = 5  # Number of similar sentences to retrieve
    similar_indices = similarities[0].argsort()[::-1][1:k+1]  # Exclude itself


    # Print similar sentences
    print(f"Similar sentences to '{input_sentence}':")
    for idx in similar_indices:
        print(f"- {sentences[idx]} (Similarity: {similarities[0][idx]})")

# Example usage
process_large_csv('/content/RAW_recipes_new.csv')
```
